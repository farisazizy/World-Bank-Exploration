# 1. Which country has the lowest death caused by nutrition in 2019? N

```python
df1 = df[['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code','2019']]
```

```python
# Checking
nutrition_cause_2019 = df1[df1['Indicator Name'].str.contains('nutrition', case=False)]
nutrition_cause_2019['Indicator Name'].unique()
```

```python
nutrition_cause_2019 = nutrition_cause_2019.dropna()
nutrition_cause_2019 = nutrition_cause_2019.sort_values('2019')
nutrition_cause_2019 = nutrition_cause_2019.head()
```

```python
nutrition_cause_2019 = nutrition_cause_2019.reset_index(drop=True)

plt.bar(nutrition_cause_2019['Country Name'], nutrition_cause_2019['2019'])
lowest_index = nutrition_cause_2019['2019'].idxmin()

bars = plt.bar(nutrition_cause_2019['Country Name'], nutrition_cause_2019['2019'])
bars[lowest_index].set_color('green')

plt.xlabel('Country Name')
plt.ylabel('Rate')
plt.title('Least 5 lowest death caused by nutrition in 2019')

for i, value in enumerate(nutrition_cause_2019['2019']):
    plt.text(i, value, str(value), ha='center', va='bottom')

plt.xticks(rotation=90)

plt.show()
```

The Answer is North Macedonia
